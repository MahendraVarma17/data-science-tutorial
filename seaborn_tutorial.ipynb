```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
```

```python
print(df.to_string)
```

```python
g=sb.FacetGrid(df,col='time')
g.map(plt.hist,'tip')
plt.show()
```

```python
g=sb.FacetGrid(df,col='day')
g.map(plt.hist,'tip')
plt.show()
```

```python
g=sb.FacetGrid(df,col='size')
g.map(plt.hist,'time')
plt.show()
```

```python
g=sb.FacetGrid(df,col='sex',hue='smoker')
g.map(plt.scatter,'total_bill',"tip",edgecolor='black')
plt.show()
```

```python
g=sb.FacetGrid(df,col='sex',hue='day')
g.map(plt.scatter,'total_bill',"tip",edgecolor='black')
plt.show()
```

```python
sb.barplot(x='sex',y='tip',hue='time',data=df)

plt.show()
```

```python
sb.regplot(x='total_bill',y='tip',color='black',data=df)

plt.show()
```

```python
sb.pairplot(df)
plt.show()
```

```python
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
sb.pairplot(df,hue="size")
plt.show()
```

```python
sb.pairplot(df,hue="size",palette='husl')
plt.show()
```

```python
sb.pairplot(df,diag_kind='kde')
plt.show()
```

```python
sb.pairplot(df,kind='reg')
plt.show()
```
